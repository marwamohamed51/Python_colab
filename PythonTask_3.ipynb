```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
"""(1)
Load the HR attrition dataset and extract one numeric column only
(MonthlyIncome) into a NumPy array. Print the array.
"""

hr=pd.read_excel('/content/Hr dataset.xlsx')
hr.info()
```

```python
MonthlyIncome=np.array(hr['Monthly Income'])
print(MonthlyIncome)
```

```python
"""(2)
Print the data type (dtype) of the MonthlyIncome array.
"""

MonthlyIncome.dtype
```

```python
"""(3)
Convert the MonthlyIncome values into a Python list and print both the
list and the NumPy array. Write one sentence describing the difference
you notice.
"""

MonthlyIncome_list=list(MonthlyIncome)
print("List")
print(MonthlyIncome_list)
print('\n')
print("Numpy Array")
print(MonthlyIncome)
print('\n')
print('The Difference is list output object values but numpy array output is values')
```

```python
"""(4)
Calculate the total MonthlyIncome using:
--Python built-in sum() on the list
--NumPy np.sum() on the array
"""

print(f"Sum of MonthlyIncome_list : {sum(MonthlyIncome_list)}")
print(f"Sum of MonthlyIncome_array : {MonthlyIncome.sum()}")
```

```python
"""(5)
Calculate the average MonthlyIncome using NumPy
"""

Avg=MonthlyIncome.mean()
print(f"Avg of MonthlyIncome : {Avg}")
print(f"Avg of MonthlyIncome : {round(Avg)}")
```

```python
"""(6)
Using simple indexing, print:
--The first 5 MonthlyIncome values
--The last 5 MonthlyIncome values
"""

print(f"The first 5 MonthlyIncome values : {MonthlyIncome[:5]}")
print(f"The last 5 MonthlyIncome values : {MonthlyIncome[-5:]}")
```

```python
"""(7)
Find the highest and lowest MonthlyIncome using NumPy functions.
"""

print(f"Highest MonthlyIncome value : {MonthlyIncome.max()}")
print(f"Lowest MonthlyIncome value : {MonthlyIncome.min()}")
```

```python
"""(8)
Increase all MonthlyIncome values by 10% using NumPy and print the
new values.
"""

MonthlyIncome_10_percent = MonthlyIncome * 1.1
print(f"MonthlyIncome_10 : {MonthlyIncome_10_percent}")
```

```python
#another way
hr['10_percent']=MonthlyIncome + MonthlyIncome*10/100
print(hr['10_percent'])
```

```python
"""(9)
Count how many employees have MonthlyIncome greater than 5000
using NumPy conditions
"""

print(f"Employees have MonthlyIncome > 5000 : {np.sum(MonthlyIncome>5000)}")
```

```python
"""(10)
Load one more numeric column (Age) into a NumPy array and print it.
"""

#Age=np.array(hr['Age'])
Age = np.array(hr.Age)
print(Age)
```

```python
"""(11)
Calculate the average Age using NumPy.
"""

avg_age=Age.mean()
print(f"Avg of Age : {round(avg_age)}")
```

```python
"""(12)
Combine Age and MonthlyIncome into one NumPy array and print its
shape.
"""

combine=np.array([Age,MonthlyIncome])
print(combine)
print(f"The of combine between Age and Monthly Income: {combine.shape}")
```

```python
""" (13)
Print how many rows and how many columns the combined array has
"""

num_rows, num_cols = combine.shape
print(f"The combined array has {num_rows} rows and {num_cols} columns")
```
