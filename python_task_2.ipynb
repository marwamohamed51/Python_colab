```python
"""
Task 1: Utility Functions:
1.Create the following functions:
    add_ numbers(a, b)
    subtract_ numbers(a, b)
    multiply_ numbers(a, b)
    divide_ numbers(a, b) Handle division by zero properly
"""
def addition(a, b):
    return a+b

def subtract_numbers(a, b):
    return a-b

def multiply_numbers(a, b):
    return a*b

def divide_numbers(a, b):
    if b == 0:
        return "Division by zero is not allowed"
    else:
        return a/b
```

```python
print(addition(16, 2))
print(subtract_numbers(16, 2))
print(multiply_numbers(0, 2))
print(divide_numbers(5, 0))
print(divide_numbers(35, 7))
```

```python
"""
Task 2: Discount Calculator:
1.Create a function:
    apply_discount(price, discount_rate)
    Inside the function:
    Use a lambda function to calculate the final price
    ask for a price from end user and Return the discounted price
"""
price = float(input("Enter the price: "))
discount_rate = float(input("Enter the discount rate: "))

discounted_price = lambda price, discount_rate: price - (price * discount_rate / 100)
print("Discounted Price: ", discounted_price(price, discount_rate))
```

```python
"""
Task 3: Console Calculator:
Build a calculator that:
Uses functions only
Supports:
    Addition
    Subtraction
    Multiplication
    Division
    Runs in a loop until the user chooses to exit
    Must Include:
    Input validation
    Clear menu
    Reusable functions
"""
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Enter operation (+, -, *, /) : ")

print("----------------------------------------")

if op == '+':
    print(f"Add Result: {num1} + {num2} = ", num1 + num2)

elif op == '-':
    print(f"Subtract Result: {num1} - {num2} = ", num1 - num2)

elif op == '*':
    print(f"Multiply Result: {num1} * {num2} = ", num1 * num2)

elif op == '/':
    if num2 == 0:
        print("Division by zero is not allowed")
    else:
        print(f"Divide Result: {num1} / {num2} = ", num1 / num2)
```

```python
#5. Nested Lists
n = int(input())
students = []
for _ in range(n):
    name = input()
    grade = float(input())
    students.append([name, grade])

grades = [student[1] for student in students]

lowest = min(grades)
second_lowest = min([g for g in grades if g != lowest])

names = [student[0] for student in students if student[1] == second_lowest]

names.sort()

for name in names:
    print(name)
```
